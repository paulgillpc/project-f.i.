```python
print(len(weights))
print(sum(weights))
```

```python
#Calculate Larry's Portfolio Returns data 
larry_return_data = larry_portfolio_df[1:]
larry_return_data.head()
```

```python
#Calculate Larry's Weighted Return Data
weighted_returns = (weights * larry_return_data)
weighted_returns.head()
```

```python
#Calculate Larry's Portfolio Returns
Larry_port_ret = pd.DataFrame(weighted_returns.sum(axis=1))
larry_port_return = Larry_port_ret.rename(columns ={ 0 :"Larry's Returns"})
larry_port_return.head()
```

```python
#Plot Larry's Performance
larry_port_return.plot(title="Larry's Weighted Portfolio Returns", figsize=(15,10))
```

```python
# Calculate and plot Cumulative Return
larry_cumulative = (1 + larry_port_return).cumprod()
larry_cumulative_plot = larry_cumulative.plot(title = "Larry's Cumulative Weighted Returns Returns", figsize =(15,10))
```
