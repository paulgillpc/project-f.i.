```python
#Calculate daily returns
aapl_returns = aapl_df.pct_change()
abnb_returns = abnb_df.pct_change()
ai_returns = ai_df.pct_change()
amc_returns = amc_df.pct_change()
an_returns = an_df.pct_change()
arkf_returns = arkf_df.pct_change()
cde_returns = cde_df.pct_change()
fcx_returns = fcx_df.pct_change()
hl_returns = hl_df.pct_change()
jmia_returns = jmia_df.pct_change()
mcfe_returns = mcfe_df.pct_change()
mgni_returns = mgni_df.pct_change()
nio_returns = nio_df.pct_change()
pep_returns = pep_df.pct_change()
pltr_returns = pltr_df.pct_change()
plug_returns = plug_df.pct_change()
pypl_returns = pypl_df.pct_change()
sfix_returns = sfix_df.pct_change()
sndl_returns = sndl_df.pct_change()
upst_returns = upst_df.pct_change()
znga_returns = znga_df.pct_change()
```

```python
# Combine stocks in Joe's portfolio
joes_stock_portfolio = pd.concat([aapl_returns, 
                                  abnb_returns, 
                                  ai_returns, 
                                  amc_returns, 
                                  an_returns, 
                                  arkf_returns, 
                                  cde_returns, 
                                  fcx_returns, 
                                  hl_returns, 
                                  jmia_returns, 
                                  mcfe_returns, 
                                  mgni_returns, 
                                  nio_returns, 
                                  pep_returns, 
                                  pltr_returns, 
                                  plug_returns,
                                  pypl_returns,
                                  sfix_returns,
                                  sndl_returns,
                                  upst_returns,
                                  znga_returns],
                                  axis="columns", join="inner")
joes_stock_portfolio
```

```python
# Assign weights on each stock
weights = [0.08121,
            0.10667,
            0.06762,
            0.06809,
            0.04630,
            0.03663,
            0.03061,
            0.04924,
            0.01608,
            0.02723,
            0.05133,
            0.04307,
            0.03404,
            0.09632,
            0.03404,
            0.05447,
            0.06809,
            0.03677,
            0.00451,
            0.04031,
            0.00738]
 
```

```python
print(len(weights))
print(sum(weights))
```

```python
#Calculate Joe's Stock Portfolio Returns data 
joes_stock_return_data = joes_stock_portfolio[1:]
joes_stock_return_data.head()
```

```python
#Calculate Joe's Stock Weighted Return Data
joes_weighted_returns = (weights * joes_stock_return_data)
joes_weighted_returns.head()
```

```python
#Calculate Joe's Stock Portfolio Returns
joes_stock_port_returns = pd.DataFrame(joes_weighted_returns.sum(axis=1))
joes_stock_portfolio_returns = joes_stock_port_returns.rename(columns ={ 0 :"Joe's  Stock Returns"})
joes_stock_portfolio_returns
```

```python
# PLot Returns over time
joes_plot = joes_stock_portfolio_returns.plot(title= "Joe's Stock Portfolio Returns", figsize = (15,10))
```

```python
# Calculate and plot Joe's Cumulative Returns
joes_cumulative = (1 + joes_stock_portfolio_returns).cumprod()
joes_cumulative_plot = joes_cumulative.plot(title = "Joe's Cumulative Returns", figsize = (15,10))
```
