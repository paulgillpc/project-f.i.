```python
# Combine stocks in Joe's portfolio
joes_stock_portfolio = pd.concat([aapl_returns, 
                                  abnb_returns, 
                                  ai_returns, 
                                  amc_returns, 
                                  an_returns, 
                                  arkf_returns, 
                                  cde_returns, 
                                  fcx_returns, 
                                  hl_returns, 
                                  jmia_returns, 
                                  mcfe_returns, 
                                  mgni_returns, 
                                  nio_returns, 
                                  pep_returns, 
                                  pltr_returns, 
                                  plug_returns,
                                  pypl_returns,
                                  sfix_returns,
                                  sndl_returns,
                                  upst_returns,
                                  znga_returns],
                                  axis="columns", join="inner")
joes_stock_portfolio
```

```python
# Assign weights on each stock
weights = [0.08121,
            0.10667,
            0.06762,
            0.06809,
            0.04630,
            0.03663,
            0.03061,
            0.04924,
            0.01608,
            0.02723,
            0.05133,
            0.04307,
            0.03404,
            0.09632,
            0.03404,
            0.05447,
            0.06809,
            0.03677,
            0.00451,
            0.04031,
            0.00738]
 
```

```python
print(len(weights))
print(sum(weights))
```

```python
#Calculate Joe's Stock Portfolio Returns data 
joes_stock_return_data = joes_stock_portfolio[1:]
joes_stock_return_data.head()
```

```python
#Calculate Joe's Stock Weighted Return Data
joes_weighted_returns = (weights * joes_stock_return_data)
joes_weighted_returns.head()
```

```python
#Calculate Joe's Stock Portfolio Returns
joes_stock_port_returns = pd.DataFrame(joes_weighted_returns.sum(axis=1))
joes_stock_portfolio_returns = joes_stock_port_returns.rename(columns ={ 0 :"Joe's  Stock Returns"})
joes_stock_portfolio_returns
```
